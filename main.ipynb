# Evolution Algorithm

>**Implementation of a Minimization Genetic Algorithm**

The project is telling us 'minimize the below function'.

**hgbat_func**

sum1 = sum(x,2);

sum2 = sum(x.^2,2);

y = (abs(sum2^2-sum1^2))^(1/2) + (0.5*sum2+sum1)/D + 0.5;

>D is the number of dimensions.

```python
import math
import random
import sys
import numpy as np
```

* max_iter = ( int ) - Maximum number of generations
* mut_rate = ( float < 1 ) - Rate of mutations
* repetitions = ( int ) - Number of experiment repetitions
* pop_size = ( int ) - Population size
* scope = [ float, float ] - Scope of possible chromosome values
* selection_size = ( int ) - Number of individuals we will compare in the selection
* fitness_function = func() - The fitness function and the function we are trying to minimize

```python
repetitions = 10
max_iter = 50
mut_rate = 0.2
pop_size = 100
scope = [-100, 100]
selection_size = 4
round_value = 10
random.seed(42)
```

```python
def fitness_function(x, D):
    
    sum1 = x + 2
    sum2 = x**2 + 2
    
    return math.sqrt(abs(sum2**2 - sum1**2)) + (0.5 * sum2 + sum1) / D + 0.5
```

```python
# for our fitness function we are using the function we are trying to minimize
def fitness_function_value(chromosome, d=5):
    fitness1 = round(fitness_function(chromosome[0], D=d), round_value)
    fitness2 = round(fitness_function(chromosome[1], D=d), round_value)
    
    if (fitness1 > fitness2):
        return fitness1
    else:
        return fitness2
```

```python
# uniform mutation
def mutate(chromosome, rate, scope):
    if random.random() < rate:
        for i in range(len(chromosome)):
            chromosome[i] = random.uniform(*scope)
    return chromosome
```

```python
# crossover single point
def crossover(chrom1, chrom2):
    # Select a random crossover point
    crossover_point = random.randint(1, len(chrom1) - 1)

    # Perform crossover by exchanging genetic material beyond the crossover point
    new_chrom1 = chrom1[:crossover_point] + chrom2[crossover_point:]
    new_chrom2 = chrom2[:crossover_point] + chrom1[crossover_point:]

    return new_chrom1, new_chrom2
```

```python
# tournament selection
def selection(population, size, dim = 5):
    competition_list = []
    #selecting the chromosomes to compete in the tournament
    while len(competition_list) < size:
        competition_list.append(random.choice(population))
    best_chromosome = None
    best_function_value = None
    #finding the best chromosome from our list of selected ones by using the fitness function
    for selected in competition_list:
        selected_value = fitness_function_value(selected, d = dim)
        if best_chromosome is None or selected_value < best_function_value:
            best_function_value = selected_value
            best_chromosome = selected
    return best_chromosome
```

```python
# chromosome fitness , number of generations
def simple_legend_func(cf = [1, 2, 3], p_label = "Generation 1", x_label = "number of generations", y_label = "chromosome fitness"):
    fig, ax = plt.subplots()

    line1, = ax.plot(cf, label=p_label, linewidth=1)
    # line2, = ax.plot([3, 2, 1], label="Line 2", linewidth=4)

    # Create a legend for the first line.
    first_legend = ax.legend(handles=[line1], loc='upper right')

    # Add the legend manually to the current Axes.
    ax.add_artist(first_legend)
    
    plt.xlabel(x_label)
    plt.ylabel(y_label)

    # Create another legend for the second line.
    # ax.legend(handles=[line2], loc='lower right')
    
    plt.show()
```

## Genetic Algorithm

```python
def genetic_algorithm(dimension = 5):
    npop_size = pop_size

    outfile = sys.stdout

    print('Algorithm started...with {} dimension'.format(dimension), file=outfile)
    for k in range(repetitions):
        best_chromosome = None
        best_fitness_value = None
        iter = 1
        
        fit_vals = np.array([])
        generation_averages = np.array([])

        # initial population generation
        pop = [[random.uniform(*scope) for i in range(2)] for j in range(pop_size)]

        # generate a new generation either until we reach the maximum number of iterations or our fitness function becomes 0
        while best_fitness_value != 0 and iter <= max_iter:
            new_pop = pop[:]

            while len(new_pop) < pop_size + npop_size:
                chrom1 = selection(pop, selection_size, dim = dimension)
                chrom2 = selection(pop, selection_size, dim = dimension)
                chrom3, chrom4 = crossover(chrom1, chrom2)
                mutate(chrom3, mut_rate, scope)
                mutate(chrom4, mut_rate, scope)
                new_pop.append(chrom3)
                new_pop.append(chrom4)

            # sort chromosomes by fitness and leave the top half for the next generation
            pop = sorted(new_pop, key=lambda chrom : fitness_function_value(chrom, d = dimension))[:pop_size]
            
            # check fitness of the current best chromosome and save if needed
            fit_val = fitness_function_value(pop[0], d = dimension)
            
            if best_fitness_value is None or best_fitness_value > fit_val:
                best_fitness_value = fit_val
                best_chromosome = pop[0]
                
            # find average
            generation_average_fitness = round(sum(map(fitness_function_value, pop)) / pop_size, round_value)
            # print('Generation {}: best chromosome fitness: {} | average fitness: {}'.format(iter, fit_val, generation_average_fitness), file=outfile)
            iter += 1
            fit_vals = np.append(fit_vals, fit_val)
            generation_averages = np.append(generation_averages, generation_average_fitness)
            
        simple_legend_func(cf = fit_vals, p_label = "Generation {}".format(str(k + 1)), y_label = "best chromosome fitness")
        simple_legend_func(cf = generation_averages, p_label = "Generation {}".format(str(k + 1)), y_label = "average fitness")
        print('Algorithm finished in {} generations.'.format(iter - 1))
        print('Best chromosome: ', [round(best_chromosome[0], 2), round(best_chromosome[1], 2)])
        print('Best fitness: {}'.format(round(best_fitness_value, 2)))
```

```python
genetic_algorithm()
```
